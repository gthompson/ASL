## 3. Plot RSAM and DRS

```python
rsamSH = obspyGT.RSAM.RSAMobj()
rsamSH.read(starttSH, endtSH, SDS_TOP, metric='mean', speed=2)
rsamSH.plot()

drsSH = obspyGT.RSAM.ReducedDisplacementObj()
drsSH.read(starttSH, endtSH, SDS_TOP, metric='median', speed=2)
drsSH.plot(type='log', linestyle='-')
```
