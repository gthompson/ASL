```python
# Convert Paul's minute files to SDS
import os, glob, obspy, datetime
# SDS_TOP is path where YYYY directories reside. SDS directory structure is like YYYY/NET/STATION/CHANNEL.D
SDS_TOP = '/Volumes/DATA/SDS'
# MINUTEFILES_TOP is path where MM directories reside. Directory structure is like MM/DD/HH
MINUTEFILES_TOP = '/Users/thompsong/Library/CloudStorage/GoogleDrive-glennthompson1971@gmail.com/Shared drives/KSC_Seismic/data/Artemis/Station_39B'
YYYY='2022' # Better if this came from directory strucuture, ie. minute files could be organized like YYYY/MM/DD/HH
monthdirs=glob.glob(os.path.join(MINUTEFILES_TOP, '??'))
for monthdir in sorted(monthdirs):
    MM = os.path.basename(monthdir)
    daydirs=glob.glob(os.path.join(monthdir, '??'))
    for daydir in sorted(daydirs):
        DD = os.path.basename(daydir)
        hourdirs=glob.glob(os.path.join(daydir, '??'))
        dayst = obspy.Stream()
        this_date = datetime.date.fromisoformat('%s-%s-%s' % (YYYY,MM,DD))
        this_jday = this_date.strftime('%j')
        print(this_date, this_jday)
        for hourdir in sorted(hourdirs):
            minute_files = glob.glob(os.path.join(hourdir,'*.ms'))
            for minute_file in sorted(minute_files):
                print('Reading %s' % minute_file)
                dayst = dayst + obspy.read(minute_file, 'mseed')
                dayst.merge(fill_value=0)
        print(dayst)
        for tr in dayst:
            sdsdir = os.path.join(SDS_TOP, YYYY, tr.stats.network, tr.stats.station, tr.stats.channel+'.D')
            os.makedirs(sdsdir, exist_ok=True)
            sdsfile = '%s.D.%s.%s' % (tr.id, YYYY, this_jday)
            tr.write(os.path.join(sdsdir, sdsfile), 'mseed')
```
